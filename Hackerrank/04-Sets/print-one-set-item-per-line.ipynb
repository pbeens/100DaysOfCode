```python
#
# Example to use Python comprehension to print a set, one item per line
#

s = set([1, 3, 2, 4, 5])
print(s)
```

```python
# This is the output I want, done using a traditional for loop

for n in sorted(s):
    print(n)
```

```python
# This is what I tried. It was explained to me that this is a generator and it 
# prints a reference to the generator

print(n for n in sorted(s))
```

```python
# This was recommended a few times, but it just prints the list, not the items 
# in the list. Note the added brackets.

print([n for n in sorted(s)])
```

```python
# This was the only recommended solution that worked. Note the use of join(). 
# Admittedly this is much more complex than my simple for loop! 

print('\n'.join([str(item) for item in sorted(s)]))
```
